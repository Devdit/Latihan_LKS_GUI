# GUI

```python
root = tk.Tk()
root.title('Obesity Clustering')
root.attributes('-fullscreen', True)

title_style = font.Font(family='Segoe UI', size = 30)
body_style = font.Font(family='Segoe UI', size = 15)

header = tk.Frame(root, bg='blue', height=20)
header.pack(side=tk.TOP, fill=tk.X)

header_title = tk.Label(header, text='Obesity Clustering', font=title_style, bg='blue', fg='white')
header_title.pack(side=tk.LEFT, padx=15)

sidebar = tk.Frame(root, bg='light blue', width=15)
sidebar.pack(side=tk.LEFT, fill=tk.Y)

main_frame = tk.Frame(root)
main_frame.pack(side=tk.TOP, fill=tk.BOTH, expand=True)



def clear_frame():
    for widgets in main_frame.winfo_children():
        widgets.destroy()

def restore_button():
    children = sidebar.winfo_children()

    for child in children[:-1]:
        child.config(bg='light blue', fg='black')

def title(Text):
    Title = tk.Label(main_frame, text=Text, font=title_style)
    Title.place(relx=.5, rely=.5, anchor=tk.CENTER)

def body(Text):
    Body = tk.Label(main_frame, text=Text, font=body_style, justify='left')
    Body.pack(side=tk.TOP, pady=10, padx=15, anchor='w')

def next(command):
    next_button = tk.Button(main_frame, text='Next Page', command=command, width=15)
    next_button.pack(side=tk.BOTTOM,anchor='s' ,pady=(0,10))

def prev(command):
    prev_button = tk.Button(main_frame, text='Previous Page', command=command, width=15)
    prev_button.pack(side=tk.BOTTOM,anchor='s' ,pady=(0,10))

def button_placeholder():
    placeholder = tk.Label(main_frame)
    placeholder.pack(side=tk.BOTTOM, anchor='s', pady=7)

def pic(image):
    picture = tk.Label(main_frame, image=image)
    picture.pack(side=tk.TOP, pady=(0,10) ,padx=15)

def page_num(num):
    page = tk.Label(main_frame, text=f'Page {num}', font=body_style)
    page.place(relx=.950, rely=.975, anchor='s')


#Importing each image that has been imported
data_head1 = tk.PhotoImage(file='data_head1.png').subsample(2,2)
data_head2 = tk.PhotoImage(file='data_head2.png')
info = tk.PhotoImage(file='info.png').subsample(2,2)


def prep():
    clear_frame()
    restore_button()
    prep_button.config(bg='blue', fg='white')

    def head_page():
        clear_frame()
        page_num('1/4')
        title('Act I: Preparation')
        next(first_page)

    def first_page():
        clear_frame()
        page_num('2/4')
        next(second_page)
        prev(head_page)
        body('Taking a look at the head of the data')
        pic(data_head1)
        pic(data_head2)
        body('As we can see, This dataset consists of mostly  numeric data, and some categorical data that we can convert to numeric \nbut first, lets check the data types')
    def second_page():
        clear_frame()
        page_num('3/4')
        next(third_page)
        prev(first_page)
        pic(info)
        body('We can see that some of the data is still object, we will convert that into numeric so we can use them in the modelling')
    def third_page():
        clear_frame()
        page_num('4/4')
        button_placeholder()
        prev(second_page)
        body('Anjayani3')

    head_page()

def prepro():
    clear_frame()
    restore_button()
    prepro_button.config(bg='blue', fg='white')

    def head_page():
        clear_frame()
        page_num('1/4')
        titles = tk.Label(main_frame, text='Act I: Preparation', font=title_style)
        titles.place(relx=.5, rely=.5, anchor=tk.CENTER)
        next(first_page)

    def first_page():
        clear_frame()
        page_num('1/3')
        title('Preprocessing')
        next(second_page)
        button_placeholder()
        body('Anjayani')
    def second_page():
        clear_frame()
        page_num('2/3')
        title('Preprocessing')
        next(third_page)
        prev(first_page)
        body('Anjayani2')
    def third_page():
        clear_frame()
        page_num('3/3')
        title('Preprocessing')
        button_placeholder()
        prev(second_page)
        body('Anjayani3')

    first_page()

def model():
    clear_frame()
    restore_button()
    model_button.config(bg='blue', fg='white')

    def first_page():
        clear_frame()
        page_num('1/3')
        title('Model')
        next(second_page)
        button_placeholder()
        body('Anjayani')
    def second_page():
        clear_frame()
        page_num('2/3')
        title('Model')
        next(third_page)
        prev(first_page)
        body('Anjayani2')
    def third_page():
        clear_frame()
        page_num('3/3')
        title('Model')
        button_placeholder()
        prev(second_page)
        body('Anjayani3')

    first_page()

def test():
    clear_frame()
    restore_button()
    test_button.config(bg='blue', fg='white')

    def first_page():
        clear_frame()
        page_num('1/3')
        title('Testing')
        next(second_page)
        button_placeholder()
        body('Anjayani')
    def second_page():
        clear_frame()
        page_num('2/3')
        title('Testing')
        next(third_page)
        prev(first_page)
        body('Anjayani2')
    def third_page():
        clear_frame()
        page_num('3/3')
        title('Preparation')
        button_placeholder()
        prev(second_page)
        body('Anjayani3')

    first_page()
        

prep_button = tk.Button(sidebar, text='Preparation', bg='light blue', width=15, bd=0, font=body_style, command=prep)
prep_button.pack(side=tk.TOP, pady=15)

prepro_button = tk.Button(sidebar, text='Preprocessing', bg='light blue', width=15, bd=0, font=body_style, command=prepro)
prepro_button.pack(side=tk.TOP, pady=15)

model_button = tk.Button(sidebar, text='Model', bg='light blue', width=15, bd=0, font=body_style, command=model)
model_button.pack(side=tk.TOP, pady=15)

test_button = tk.Button(sidebar, text='Testing', bg='light blue', width=15, bd=0, font=body_style, command=test)
test_button.pack(side=tk.TOP, pady=15)

exit_button = tk.Button(sidebar, text='EXIT', bg='red',fg='white' ,width=15, bd=0, font=body_style, command=root.destroy)
exit_button.pack(side=tk.BOTTOM, pady=15)

root.mainloop()
```
